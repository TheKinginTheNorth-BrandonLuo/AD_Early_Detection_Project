```python
import math
import numpy as np
from sklearn import metrics
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
def NMI(A,B):
    total = len(A)
    A_ids = set(A)
    B_ids = set(B)
    #mutual information
    MI = 0
    eps = np.finfo(float).eps
    for idA in A_ids:
        for idB in B_ids:
            idAOccur = np.where(A==idA)
            idBOccur = np.where(B==idB)
            idABOccur = np.intersect1d(idAOccur,idBOccur)
            px = 1.0*len(idAOccur[0])/total
            py = 1.0*len(idBOccur[0])/total
            pxy = 1.0*len(idABOccur)/total
            MI = MI + pxy*math.log(pxy/(px*py)+eps,2)
    #standatd mutual information
    Hx = 0
    for idA in A_ids:
        idAOccurCount = 1.0*len(np.where(A==idA)[0])
        Hx = Hx - (idAOccurCount/total)*math.log(idAOccurCount/total+eps,2)
    Hy = 0
    for idB in B_ids:
        idBOccurCount = 1.0*len(np.where(B==idB)[0])
        Hy = Hy - (idBOccurCount/total)*math.log(idBOccurCount/total+eps,2)
    MIhat = 2.0*MI/(Hx+Hy)
    return MIhat

if __name__ == '__main__':
    dt = pd.read_csv('added_other_features_final.csv')
    dt = dt.drop(columns=['date','year','getSUBTLWordScores'],axis=1)
    A = dt['CC']
    B = dt['DT']
    C = dt['FW']
    D = dt['PRON_rate']
    E = dt['IN']
    F = dt['JJ']
    G = dt['MD']
    H = dt['NN']
    I = dt['NNS']
    J = dt['NNP']
    K = dt['PDT']
    L = dt['PRP']
    M = dt['PRP$']
    N = dt['RB']
    O = dt['RP']
    P = dt['VB']
    Q = dt['VBD']
    R = dt['VBP']
    S = dt['VBZ']
    T = dt['WP']
    U = dt['WRB']
    V = dt['$']
    W = dt['SYM']
    X = dt['ADV']
    Y = dt['NOUN']
    Z = dt['VERB']
    AA = dt['PRON']
    AB = dt['PRP_NOUN_Ratio']
    AC = dt['NOUN_Ratio']
    AD = dt['PRP_Ratio']
    AE = dt['word_freq_rate']
    AF = dt['word_freq_rate_Nsw']
    AG = dt['verb_freq_rate']
    AH = dt['Noun_rate']
    AI = dt['PRON_rate']
    AJ = dt['ADV_rate']
    AK = dt['H_S']
    AL = dt['BM']
    AM = dt['TTR']
    AN = dt['SICH']
    AO = dt['A_R_I']
    AP = dt['F_K_R']
    AQ = dt['getFamiliarityScore']
    AR = dt['getConcretenessScore']
    AS = dt['getImagabilityScore']
    AT = dt['getAoaScore']

    print("The mutual information of PRP_NNOUN_Ratio is:",NMI(C,D))
    print(metrics.normalized_mutual_info_score(C,D))
```
